```python
import pandas as pd
df=pd.read_excel("/content/US_Customer.xlsx")
df.head()
```

```python

print("first 5 Rows")
print(df.info())

print("\n Non Null Data Counts")
print(df.describe())

print("\n Missing Values per column")
print(df.isnull().sum())
df.head()
```

```python
numdata=df.select_dtypes(include=["int64","float64"]).columns.tolist()
print(numdata)
catdata=df.select_dtypes(include=["object","category"]).columns.tolist()
print(catdata)
df.head()
```

```python
df.columns=df.columns.str.strip().str.replace(' ','_').str.lower()

print(df.columns[df.columns.duplicated()])
# remove duplicate column names
df = df.loc[:, ~df.columns.duplicated()].copy()

df.head()
```

```python

df['transactiondate'] = pd.to_datetime(
    df['transactiondate'],
    errors='coerce',           # invalid formats -> NaT
    infer_datetime_format=True # auto-detect format
)


df.head()
```

```python
invalid_dates = df[df['transactiondate'].isna()]
print(" Unparseable date rows:")
print(invalid_dates[['transactiondate']])
```

```python
print(df['transactiondate'].unique()[:10])
```

```python
df['transactiondate'] = df['transactiondate'].astype(str).str.strip()
df['transactiondate'] = df['transactiondate'].str.extract(r'(\d{2}/\d{2}/\d{4})')[0]
df['transactiondate'] = pd.to_datetime(df['transactiondate'], errors='coerce')
df['transactiondate'] = pd.to_datetime(df['transactiondate'], format='%d/%m/%Y', errors='coerce')
df['transactiondate'] = df['transactiondate'].dt.date
df.head()
```

```python
print(df)
```

```python
from dateutil import parser

def parse_date_safe(x):
    try:
        return parser.parse(str(x), dayfirst=True)  # favor dd/mm/yyyy
    except Exception:
        return pd.NaT   # unparseable  NaT

df["transactiondate"] = df["transactiondate"].apply(parse_date_safe)
```

```python
df["transactiondate_raw"] = df["transactiondate"]  # keep original
df["transactiondate"] = df["transactiondate"].apply(parse_date_safe)
df.head()
```

```python
df["joindate"]=pd.to_datetime(df["joindate"],dayfirst=True,errors="coerce")
#df['transactiondate']=pd.to_datetime(df["transactiondate"],dayfirst=True,errors="coerce")
df['joindate']=df['joindate'].dt.strftime('%d/%m/%Y')
#df['transactiondate']=df['transactiondate'].dt.strftime('%d/%m/%Y')
df.head(25)
```

```python
df.to_csv("Cleaned_Customer_data.csv",index=False)
```
