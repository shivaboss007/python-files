```python
import pandas as pd
df1=pd.read_csv("/content/index_1.csv")
df2=pd.read_csv("/content/index_2.csv")
df3=pd.concat([df1,df2],ignore_index=True)
```

```python
great_30=df1[df1['money']>30]
print(great_30)
```

```python
card_trasactions=df1[df1['cash_type']=="card"]
print(card_trasactions)
```

```python
coffee_filter=df2[df2['coffee_name'].isin(["Latte","Americano"])]
print(coffee_filter)
```

```python
cf=df2[(df2['coffee_name']=="Latte")|(df2['coffee_name']=="Americano")]
print(cf)
```

```python
from datetime import datetime
df1['datetime']=pd.to_datetime(df1['datetime'],errors="coerce")
dt=df1[df1['datetime']>"2024-03-01 12:00:00"]
print(dt)
```

```python
ct=df1[df1['card']=='ANON-0000-0000-0002']
print(ct)
```

```python
fil_coffee=df3.groupby("coffee_name")['money'].sum().sort_values(ascending=False).reset_index()
print(fil_coffee.head())
```

```python
numberoftransactions=df3.groupby('cash_type')['cash_type'].count()
print(numberoftransactions)
```

```python
df1cash=df1['cash_type'].value_counts().reset_index(name="count1")
#df1cash.columns=['cash_type', 'count1']
print(df1cash)
df2cash=df2['cash_type'].value_counts().reset_index(name="count2")
#df2cash.columns=['cash_type', 'count2']
print(df2cash)
df3cash=pd.merge(df1cash,df2cash,on='cash_type',how='outer').fillna(0)
print(df3cash)
```

```python
avgspendoncoffee=df1.groupby("coffee_name")['money'].mean().sort_values(ascending=False)
print(avgspendoncoffee)
```

```python
highestrevenue=df2.groupby('coffee_name')['money'].sum().sort_values(ascending=False).reset_index(name="Revenue")
print(highestrevenue.head())
```

```python
df1['datetime']=pd.to_datetime(df1['datetime'],errors='coerce')
df2['datetime']=pd.to_datetime(df2['datetime'],errors='coerce')
df1_card=df1[df1['cash_type']=="card"][['cash_type','money']]
df2_card=df2[df2['cash_type']=="card"][['cash_type','money']]
mergedf=pd.merge(df1_card,df2_card,how='outer')
#print(df1_card)
#print(df2_card)
print(mergedf)
```

```python
mdf=df3[df3['cash_type']=='card'][['cash_type','money']].reset_index(drop=True)
print(mdf)
```

```python
innerjcoff=pd.merge(df1,df2,on='coffee_name', how='inner',suffixes=(1,2))
print(innerjcoff.head())
print(len(innerjcoff))
```

```python
perdayrevenue=df3.groupby('date')['money'].sum().reset_index()
print(perdayrevenue)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
sns.barplot(
    data=fil_coffee.head(10),
    x="coffee_name",
    y='money',
    palette='viridis',

)
plt.title("Bar Chart")
plt.xlabel("Coffee name",fontsize=15)
plt.ylabel("Revenue",fontsize=15)
plt.xticks(rotation=45, ha='right')
plt.show()
```
