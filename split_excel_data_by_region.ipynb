```python
# Import the pandas library for data manipulation
import pandas as pd
# Read the Excel file 'Dataset1.xlsx' into a pandas DataFrame
df=pd.read_excel("/content/Dataset1.xlsx")
# Display the DataFrame (implicitly shows the head and tail by default in Colab)
df
```

```python
# Get all unique values from the 'Region' column of the DataFrame
df['Region'].unique()
```

```python
# Iterate through each unique region found in the 'Region' column
for reg in df['Region'].unique():
  # Print the current region name
  print(reg)
```

```python
# Create an ExcelWriter object to save data to 'Splited_Data.xlsx' using the 'xlsxwriter' engine
write=pd.ExcelWriter("Splited_Data.xlsx",engine='xlsxwriter')
# Loop through each unique region in the DataFrame
for reg in df['Region'].unique():
  # Filter the original DataFrame to create a new DataFrame for the current region
  newdf=df[df['Region']==reg]
  # Write the filtered DataFrame to a new sheet in the Excel file, named after the region
  # index=False prevents writing the DataFrame index to the Excel sheet
  newdf.to_excel(write,sheet_name=reg,index=False)

# Save and close the Excel file. This is crucial for writing the data to disk.
write.close()
```
